I dette tilfellet virker varianten med __read()__ mest tungvint av de tre
- må selv splitte to ganger, først på '\n', så på ' '
- men __kan__ være nyttig hvis vi ønsker å gjøre en massiv replace-operasjon
    - eksempel, hvis fila hadde norsk komma i tallene, vil bytte til punktum
        - enklere å gjøre det da med hele teksten som en streng, enn etter at splittet opp
    - eller: fila er delt inn i bolker med et skilletegn, f.eks. '@'
        - deretter er hver bolk delt i linjer
        - ønsker da å splitte på '@' først, mens for... / readlines() vil splitte på \n først
        
Et annet case for __read()__ kan være tidsbruk
- fillesing vanligvis tregere enn operasjoner på variable i primærminnet
- derfor raskere å gjøre all fillesing i ett jafs, heller enn litt og litt i løkke

## Del 3: Lesing fra fila, inkludert unntaksbehandling
Mye kan gå galt under lesing fra fil. F.eks.
- fila kan ha blitt fjerna, flytta eller endra navn på
    - vil da få FileNotFoundError
- det kan mangle data på fila
    - f.eks. at atomvekt mangler for ett av atomene
    - vil da få IndexError fordi liste\[3\] ikke fins
- data på fila kan være på feil format
    - f.eks. et flyttall oppgitt med komma, ikke punktum
    - vil da få ValueError

Dette kan du prøve ut ved å
- legge inn slike feil i fila
- kjøre koden for å lese fra fila på nytt

```python
def les_grunnstoff_fra_fil(filnavn):
    '''Leser tekstfil med data om grunnstoff: atomnr navn symbol atomvekt,
       returnerer ei 2d liste på formen [[symbol, navn, atomnr, atomvekt],...]
    '''
    data = []
    try:
        with open(filnavn) as fil:
            for linje in fil:               
                liste = linje.split()
                atomnr = int(liste[0])
                atomvekt = float(liste[3])
                data.append([liste[2], liste[1], atomnr, atomvekt])
    except:
        print('Noe gikk galt under lesing fra fil!')
    return data

grunnstoff = les_grunnstoff_fra_fil('grunnstoff.txt')
print(f'Sluttresultat:\n{grunnstoff}')
```

Det er mulig (men tungvint) å lage if-setninger for å
- sjekke at fila fins før vi åpner den
- sjekke at alle linjer på fila har nok data
- sjekke at alle data er på rett format

Enklere å sette alt som har med fila å gjøre i ei try-blokk, som vist nedenfor.
- return-setninga trenger ikke være i try, den kan vi gjøre uansett
- (vil returnere tom dictionary hvis lesing av fil ikke lykkes)

```python
def les_grunnstoff_fra_fil(filnavn):
    '''Leser tekstfil med data om grunnstoff: atomnr navn symbol atomvekt,
       returnerer ei 2d liste på formen [[symbol, navn, atomnr, atomvekt],...]
    '''
    data = []
    try:
        with open(filnavn) as fil:
            for linje in fil:               
                liste = linje.split()
                atomnr = int(liste[0])
                atomvekt = float(liste[3])
                data.append([liste[2], liste[1], atomnr, atomvekt])
    except:
        print('Det er noe feil med fila!') 
        
    return data

grunnstoff = les_grunnstoff_fra_fil('grunnstoff.txt')
print(f'Sluttresultat:\n{grunnstoff}')
```

```python
def les_grunnstoff_fra_fil(filnavn):
    '''Leser tekstfil med data om grunnstoff:
       atomnr navn symbol atomvekt,
       returnerer en dictionary på formen
       {symbol: [navn, atomnr, atomvekt]}'''
    d = {} # oppretter tom ordbok (dictionary)
    try:
        with open(filnavn) as fil:
        with open(filnavn) as fil:
            for linje in fil:               
                liste = linje.split()
                atomnr = int(liste[0])
                atomvekt = float(liste[3])
                data.append([liste[2], liste[1], atomnr, atomvekt])
    except FileNotFoundError:
        print('Fant ikke fila, sjekk om filnavn og katalogplassering er korrekt')
    except IndexError:
        print('Ei linje på fila hadde for lite data')
    except ValueError:
        print('Data som skulle konverteres til tall var på feil format')
    return d 

grunnstoff = les_grunnstoff_fra_fil('grunnstoff.txt')
print(grunnstoff)
```

## Oppsummering
Bruk av fil
- må først åpne fila, typisk med __with open(filnavn) as filvariabel:__

Lesing:
- for linje in fil: ... leser ei og ei linje
- readlines() ... leser hele fila inn i ei liste av strenger, splittet på linjeskift
- read() ... leser hele fila som én lang strengvariabel (må eventuelt selv splitte på \n)

Skriving:
- write() - skriver en streng
    - må selv passe på å sette \n bakerst for å få med linjeskift
    - eller joine ei hel liste av strenger på \n og skrive hele fila i ett jafs
- writelines() - skriver ei hel liste av strenger
    - setter automatisk inn \n mellom hver streng
    
Unntaksbehandling:
- hvis det er stor fare for at noe går galt med bruk av fil
- lurt å sette koden i try-blokk
- except-delen angir hva som alternativt skal skje hvis programmet feiler
    - kan enten ha bare én except som tar alle typer feil
    - eller ulike except for ulike feil (f.eks. FileNotFoundError, IndexError, ValueError...)
