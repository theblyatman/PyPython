# Starter med et eksempel

```python
def f(x, y):
    w = [0, 0, 0]
    x[0] = 7.7
    
v = [1, 2, 3]
w = [4, 5, 6]
f(v, w)
print("v etter f()", v)
print("w etter f()", w) 
```

- __v__ har endret verdi, __w__ ikke
- HVORFOR?
    - det ser da ut som f() endrer w
    - mens den ikke gjør noe med v ?
- Under forklares hva som skjer steg for steg i koden
    - #nr viser rekkefølge

```python
def f(x, y):                   #1 funksjon f defineres;    #4 x, y ble kalt med v, w; viser til samme lister
    w = [0, 0, 0]                                          #5 oppretter lokalt navn w (ANNET listeobjekt)
    x[0] = 7.7                                             #6 x (samme som v) endrer første element til 7.7
    
v = [1, 2, 3]                  #2 oppretter v, listeobjektet [1,2,3]
w = [4, 5, 6]                  #3 oppretter w, listeobjektet [4,5,6]
f(v, w)                        #4 kaller f(v,w). Dette betyr at inni f blir x samme liste som v, y som w
print("v etter f()", v)                                    #7 print er, v har blitt [7.7,2,3] pga endring i f
print("w etter f()", w)                                    #8 printer, w ikke endret (kodelinje 2: lokal i f)
```

(Her skulle det ha vært et par figurer som illustrerer minnelokasjoner osv, men det er ikke så farlig)

Vi kan illustrere dette videre ved å printe minnelokasjonene til ulike variable

Som vi ser av minneadressene, er
- x, y inni f SAMME listeobjekt som v, w globalt i skriptet
    - dette er det som skjer ved parameteroverføring til funksjoner:
    - overfører referaser til objektene
- w inni f er derimot et NYTT listeobjekt som vi oppretter der og da
    - fordi w er et nytt navn som opprettes lokalt inni f
    - dette er det som skjer når vi oppretter en variabel med __navn = ...__

```python
def f(x, y):
    print("x, y inni f:", id(x), id(y))
    w = [0, 0, 0]
    print("og w inni f:", id(w))
    x[0] = 7.7
    print("og x inni f:", id(x))
    
v = [1, 2, 3]
w = [4, 5, 6]
print("v,w globalt:", id(v), id(w))
f(v, w)
print("v etter f()", v)
print("w etter f()", w)
```

## Små endringer kan lage en motsatt situasjon
I det neste eksemplet bytter vi om ved å
- gjøre en annerledes tilordning av w inni f
    - bruker indeksparentes på venstre side av __=__
    - dermed gjør f bruk av den globale variabelen w
- gjøre en annerledes tilordning av x
    - bruker IKKE indeksparentes på venstre side
    - dermed omdirigeres x til et nytt listeobjekt
        - i stedet for å mutere det gamle objektet

```python
def f(x, y):
    w[:] = [0, 0, 0]    # var i forrige versjon w = [0, 0, 0]
    x = [7.7, 2, 3]     # var i forrige versjon x[0] = 7.7
    
v = [1, 2, 3]
w = [4, 5, 6] 
f(v, w)
print("v etter f()", v)     # ble i forrige versjon [7.7, 2. 3]
print("w etter f()", w)     # ble i forrige versjon uendret [4,5,6]
```

## Lokale variable kan ha effekt globalt...
Hvis de gjør noe som påvirker en global variabel

Nytt eksempel, nå med lokal variabel s
- ser at v blir endret, 
- men w ikke. 

Hvorfor?

```python
def f(x, y):
    s = []
    t = []
    s = x
    t[:] = y
    s[0] = 7.7
    t[0] = 7.7
    
v = [1, 2, 3]
w = [4, 5, 6]
f(v, w)
print("v etter f()", v)
print("w etter f()", w)
```

Nedenfor vises minneadresser til variablene

Hovedpoeng:
- s og t opprettes som tomme lister (linje 3, 4)
    - dette blir lokale variable, separate minneadresser
- endringen s=x gjør imidlertid at s blir SAMME objekt som x
- t[:]=y gjør derimot at t beholder sin eksisterende minnelokasjon
    - men innholdet fra y muteres inn i stedet for nåværende innhold

```python
def f(x, y):
    print("x og y inni f:", id(x), id(y))
    s = []
    t = []
    print("s og t som []:", id(s), id(t))
    s = x
    t[:] = y
    print("s og t endret:", id(s), id(t))
    s[0] = 7.7
    t[0] = 7.7
    print("s og t sist f:", id(s), id(t))
    
v = [1, 2, 3]
w = [4, 5, 6]
print("v og w global:", id(v), id(w))
f(v, w)
print("v etter f()", v)
print("w etter f()", w)
```

## Argument -> parameter
Blir samme objekt KUN hvis
- argumentet er bare et variabelnavn
- hvis det er en verdi eller et større beregningsuttrykk
    - blir parameter et annet objekt enn argumentet

Dette er også tilsvarende for vanlige tilordninger:
- __a = b__  fører til at a blir samme objekt som b
- __a = b + 0.0__ fører til at a blir et annet objekt

```python
def f(x, y):
    print("x og y:", id(x), id(y))
    s = [0, 0, 0]
    x[0] = 7.7
    
v = [1, 2, 3]
w = [4, 5, 6]
print("v og w:", id(v), id(w))
f(list(tuple(v)), w)
print("v etter f()", v)
print("w etter f()", w)
```

## Oppsummering
Når en funksjon kalles
- hvis argumentet i kallet er et variabelnavn,
    - vil parameternavnet inni funksjonen vise til samme objekt
- hvis argumentet er en ren verdi eller et beregningsuttrykk
    - vil parameteren ikke vise til et globalt objekt

Tilordninger:
- variabel1 = variabel2
    - gjør at variabel1 viser til samme objekt som variabel2 allerede viste til
    - hvis vi deretter muterer variabel1 vil vi dermed også endre variabel2
- variabel1 = beregning på variabel 2
    - gjør at variabel1 blir et nytt objekt
- variabel1[indeksuttrykk] = ...
    - gjør at variabel1 muteres i sin nåværende minnelokasjon
    - her må variabel1 allerede eksistere på forhånd
