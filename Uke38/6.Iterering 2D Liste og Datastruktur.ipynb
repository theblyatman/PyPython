# Iterere gjennom 2d liste, lage 2d datastruktur
Eksempel her:
- iterere i dobbel løkke gjennom 2d liste (liste av lister)
- lage ei liste av tupler

Har ei 2d liste, f.eks.

```python
liste = [[3, 2, 5, 1],
         [7, -1, -4, 0],
         [-3, 0, 4, 1, 2]]
```

Vil lage en funksjon som analyserer lista og
- returnerer ei liste av tupler
- der hvert tuppel har innhold (s, p, n), der
    - s er summen av positive tall i tilsvarende indre liste
    - p er produktet av tall ulik null i tilsvarende indre liste
    - n er antall element i tilsvarende indre liste
    
Med vår eksempelliste skal returnert liste av tupler bli:

[ (11, 30, 4), (7, 28, 4), (7,-24, 5) ]


```python
def analyser_liste(lst):
    '''In: lst - a list of lists containing numbers
    Returns a list of tuples (s,p,a), where each tuple has info
    s : sum of positive elements in one inner list
    p : product of non-zero elements in one inner list
    n : number of elements in the inner list'''
    result = [ ]
    for i in range(len(lst)):
        s = 0
        p = 1
        n = len(lst[i])
        for j in range(n):
            if lst[i][j] > 0:
                s += lst[i][j]
            if lst[i][j] != 0:
                p *= lst[i][j]
        result.append((s, p, n)) # NB: trenger dobbel parentes
    return result

print(analyser_liste(liste))
```

Vi endrer ikke noe i lista vi får inn
- trenger derfor ikke iterere på indekser
- kan også iterere på elementverdi

```python
def analyser_liste(lst):
    '''In lst - a list of lists containing numbers
    Returns a list of tuples (s,p,a), where each tuple has info
    s - sum of positive elements in one inner list
    p - product of non-zero elements in one inner list
    n - number of elements in the inner list'''
    result = [ ]
    for rad in lst:
        s = 0
        p = 1
        n = len(rad)
        for tall in rad:
            if tall != 0:
                p *= tall
                if tall > 0:
                    s += tall
        result += [(s, p, n)]  # innerste parentes er tuppelparentesen
    return result

print(analyser_liste(liste))
```

Kombinasjon av itereringsmåter er også mulig

For å få litt mer enn bare tall inn her, så kan vi se for oss en konkurranse der vi har ulike personer som har samlet poeng. En todimensjonal liste der en eksempelvis har hatt andre funksjoner som har lagt dette materialet inn. Hver subliste er én person. Første element i sublisten er personens navn, de andre elementene representerer poeng fått for hver gang de deltok. Hver person kan delta et ulikt antall ganger. Poenget er å lage en graf som viser hvor mange poeng hver person har fått totalt.

```python
import matplotlib.pyplot as plt

poeng = [["Mathias", 3, 2, 5, 1],
         ["Mina", 7, -1, -4, 0],
         ["Morten minstemann", -3, 0, 4, -1, -2, -14, ]] # He is small!

'''
In: two-dimensional list, each inner list contains a name (first) and a varying set of points for that person.
Out: A tuple containing two tuples, the first contains the names in the order they appear. The second contains
the total of points each player has gotten, in the same order as the names.
'''
def finn_total_per_person(lst):
    print("hei?")
    navn = []
    total_poeng = []
    for person_data in lst:
        navn.append(person_data[0])
        total_poeng.append(sum(person_data[1:]))
    return (tuple(navn), tuple(total_poeng))

personer, totalpoeng = finn_total_per_person(poeng)

'''
Ved å hente inn funksjonen for plotting fra H-11 så kan vi raskt få på plass 
'''

def plott_søyle(tekster, tall, tittel): # Denne er hentet direkte fra H-11
    fig, ax = plt.subplots()
    ax.bar(tekster, tall)
    plt.title(tittel)
    plt.show()

plott_søyle(personer, totalpoeng, "Poeng per person")
```
