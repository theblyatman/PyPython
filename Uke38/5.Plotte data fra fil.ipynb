# Plotte data fra fil
Typisk framgangsmåte:
- Les data fra fila
    - f.eks. med numpy.loadtxt()
- Bruk data fra arrayet til å plotte, f.eks.
    - en rad som x-verdier
    - annen rad som y-verdier
- Vis plottet på skjermen i ønsket format

## Eksempel 1: Kombinere alle rader med alle (unntatt seg selv)
- med to rader blir det to trekanter
    - rad 0 som x, rad 1 som y
    - rad 1 som x, rad 0 som y
- med tre rader blir det seks:
    - rad 0, 1; rad 0, 2; rad 1, 0; rad 1, 2; rad 2, 0; rad 2, 1
    
Fremgangsmåte: Gå i dobbel løkke
- ytterste løkke finner rada som skal brukes som x-verdier
- innerste finner rada for y-verdier
- nøsting av løkkene gjør at x og y kombineres på flere ulike måter

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polygons(filename):
    fig, ax = plt.subplots()
    # kombinere alle rader med alle (unntatt seg selv)
    data = np.loadtxt(filename)
    for i in range(len(data)):
        for j in range(len(data)):
            if i != j:
                ax.plot(data[i], data[j])
    plt.show()
    
plot_polygons("triangles.txt")
```

## Ønsker å legge til "legend"
så figuren viser hvilken trekant bruker hvilke rader i fila

Fremgangsmåte:
- lage ei tom liste før løkkene
- inni innerste løkke
    - legge til en legend-tekst hver gang vi plotter
- etter løkkene: lage legend med lista av legend-tekster

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polygons(filename):
    fig, ax = plt.subplots()
    # kombinere alle rader med alle (unntatt seg selv)
    data = np.loadtxt(filename)
    legend_list = [ ]
    for i in range(len(data)):
        for j in range(len(data)):
            if i != j:
                ax.plot(data[i], data[j])
                legend_list.append(f'x: data[{i}] ; y: data[{j}]')
    ax.legend(legend_list)
    plt.show()
    
plot_polygons("triangles.txt")
```

## Eksempel 2: Kombinere alle med alle, men alltid x over y
- med to rader blir det kun 1 trekant
    - rad 0 som x, rad 1 som y
- med tre rader blir det 3 trekanter:
    - rad 0, 1; rad 0, 2; rad 1, 2

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_polygons(filename):
    fig, ax = plt.subplots()
    # kombinere alle rader med alle, men alltid x over y
    data = np.loadtxt(filename)
    legend_list = [ ]
    for i in range(len(data)-1):
        for j in range(i+1,len(data)):
            ax.plot(data[i], data[j])
            legend_list.append(f'x:data[{i}] y:data[{j}]')
    ax.legend(legend_list)
    plt.show()
    
plot_polygons("triangles.txt")
```

## Eksempel 3: Kombinere alle partall (x) med oddetall (y)
- med fire rader blir det 4 trekanter
    - rad 0 som x, rad 1 som y
    - rad 0 som x, rad 3 som y
    - rad 2 som x, rad 1 som y
    - rad 2 som x, rad 3 som y

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_polygons(filename):
    fig, ax = plt.subplots()
    # kombinere par-indeks (x) med odde-indeks (y)
    data = np.loadtxt(filename)
    legend_list = [ ]
    for i in range(0,len(data),2):
        for j in range(1,len(data),2):
            ax.plot(data[i], data[j])
            legend_list.append(f'x:data[{i}] y:data[{j}]')
    ax.legend(legend_list)
    plt.show()
    
plot_polygons("triangles.txt")
```
