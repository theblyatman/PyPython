## Konvertering __fra__ boolske verdier
Generelt tilsvarer True tallet 1, og False tallet 0

```python
# Eksempel
print(int(True)) # 1
print(int(False)) # 0
print(float(True)) # 1.0
print(float(False)) # 0.0
print(str(True)) # 'True'
print(str(False)) # 'False'
```

## Konvertering til boolske verdier
Dette kan gjøres med funksjonen __bool()__

Følgende blir __False__ med bool() av de datatypene vi har sett hittil:
- tallene 0 og 0.0
- helt tom streng ""
- helt tom liste

```python
# Eksempel
a = 0
ba = bool(a) # blir False

x = 0.0
bx = bool(x) # blir False

tekst = ''
bt = bool(tekst) # blir False

liste = []
bli = bool(liste) # blir False
```

__Alle andre verdier__ av int, float, string, list blir __True__

```python
a = -1                # -1 blir True selv om også 1 er True, kun 0 blir False
ba = bool(a)

x = 0.001             # nær 0.0 er ikke nok, kun 0.0 blir False
bx = bool(x)

tekst = 'False'       # selv 'False' blir True, er ikke en tom streng
bt = bool(tekst)

liste = [False]       # selv ei liste med kun ett element, som er False, blir True
bli = bool(liste)     # er ikke ei tom liste
```
