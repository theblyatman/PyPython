# Indeksering av strenger i Python
Som vi har vært inne på før, kan vi finne enkelttegn i strenger ved hjelp av indeks.

Indeks i Python starter på \[0\], dvs.
- indeks 0 gir første tegn i strengen, 1 gir andre tegn, 2 tredje tegn, osv.

Vi kan også indeksere bakfra:
- indeks -1 gir siste tegn, -2 nest siste tegn, -3 tredje siste tegn, osv.

```python
## Eksempel
tekst = 'Python'
print(tekst[0], tekst[3], tekst[-1], tekst[-2])
# P h n o
```

## Sammenligning av enkeltbokstaver i Python
Vi kan sammenligne bokstaver med relasjonsoperatorene ==, !=, <, <=, >=, >

Merk:
- <, <=, >=, > sammenligner basert på tegnets Unicode-verdi
- dette funker som alfabetisk sammenligning HVIS

```python
# Eksempel
print("a" == "a", "a" == "A")    # Første blir True, andre False
print("a" < "b", "b" < "C")      # første True, andre False (store har lavere Unicode)
print("æ" < "ø", "ø" < "å", "æ" < "å")
```

Som vi ser, gir linje 4 over et rart resultat:
- vi skulle hatt True True True ut fra alfabetiske regler
- men to siste blir False, fordi å har lavere Unicode-verdi enn æ, ø
- samme problem vil finnes for stor Å vs. Æ, Ø

```python
# Eksempel, indeksering og sammenligne enkelttegn
t = "oslo"
u = "solo"
print(t[1] == u[0])
# True
print(t[0] == t[-1])  # sjekker om streng begynner og slutter på samme tegn
# True
print(t[0] >= u[0])
# False
```

# Sammenligning av hele strenger
Hele strenger kan også sammenlignes med relasjonsoperatorer

== vil bli True bare hvis hele strengen er identisk
- og store/små tegn må stemme overens i begge

```python
"Python" == 'Python' # True
"Python" == "python" # False
"Python" != "python" # True
```

Alfabetisk sammenligning med >, <, vil virke som forventet
- starter på fremste tegn
- sammenligner ett og ett tegn til den finner en forskjell
- men igjen: kan feile hvis ett ord har stor bokstav hvor det andre har liten
- eller med norske tegn, Å/å vs Æ/æ, Ø/ø

```python
"Python" < "Pythonn"  # True: identiske ord, inntil det ene slutter, det regnes da som minst
"Python" < "Pythox"  # True, x er lenger i i alfabetet enn n
"python" < "Pythox"  # blir False, p > P i Unicode-verdi
"Storå" > "Storø"  # skulle vært True alfabetisk, blir False da å har for lav verdi
```

## Finne Unicode-verdi for symbol
Funksjonen __ord()__ kan brukes for å finne Unicode-verdi for et symbol

```python
# Eksempel
ord('A'), ord('Z'), ord('a'), ord('z')
# (65, 90, 97, 122)
```

```python
ord('Æ'), ord('Ø'), ord('Å'), ord('æ'), ord('ø'), ord('å')
# (198, 216, 197, 230, 248, 229)
```

## Finne symbol for en Unicode-verdi

Funksjonen __chr()__ gir symbol for en Unicode-verdi

```python
# Eksempel
chr(229), chr(8734)
# ('å', '∞')
```
