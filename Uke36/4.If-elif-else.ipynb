# if- elif- else-setninger
Tidligere i dette kurset: enkle program der alle kodelinjer skulle utføres uansett. 

Ofte er det ikke slik, koden trenger å ta __beslutninger__
- basert på brukerens handlinger eller verdien av variable

## Syntaks og virkemåte, if-setning
if _logisk uttrykk_ __:__ <br>
&nbsp;  &nbsp;&nbsp;      _handling(er)_

- _logisk uttrykk_ må kunne regnes ut / konverteres til en boolsk verdi
- if-linja __må__ avsluttes med kolon :
- _handlinger_ må være __rykket inn__ en marg
    - viser at kodelinje(ne) tilhører if-testen
    - blir kun utført hvis _logisk uttrykk_ er __True__
    
Vi kan ha flere påfølgende linjer på innrykk
- i så fall tilhører alle disse linjene if-setningen
- som vist nedenfor; her blir to printer ikke utført fordi testen er False

```python
x = -5
print("Før if-setninga")
if x >= 0:
    print("Inni if-setninga")
    print("Hva med denne?")
print("Etter if-setninga")
```

## if-else-setning
Den enkle if-setninga er slik at
- vi gjør noe kun hvis testen er True
- gjør ingenting hvis den er False

I mange tilfeller skal vi gjøre en alternativ handling hvis testen feiler.
Slike alternative handlinger får vi gjort ved å legge til __else:__

```python
# Eksempel: kvadratrot
# Kjør dette to ganger, en med positivt og en med negativt tall
x = float(input("Gi inn et tall for kvadratrotberegning: "))
if x >= 0:
    print("Kvadratrota blir", x**0.5)
else:
    print("Kan ikke regne kvadratrot av negativt tall")
print("Ha en fin dag!")
```

```python
# Eksempel: kvadratrot, lager det som funksjon heller enn med input()
def rot(x):
    if x >= 0:
        return x ** 0.5
    else:
        return "Rot av negativt tall går ikke"
    
print(rot(3))
print(rot(-4))
```

if-setninga tar et valg basert på verdien som kommer inn for parameteren x. 
- Kallet kvadratrot(3) fører oss til if-delen
- Kallet kvadratrot(-4) fører oss til else-delen.

## Nøstede if-setninger
Vi kan putte if-setninger inni if- og else-delene av andre if-setninger
- dvs har flere if-else-setninger inni hverandre
- dette kalles __nøstede if-setninger__.

Eksempel: Utvider kvadratrotfunksjonen slik at den både skal returnere selve svaret og en forklaring:

```python
def rot(x):
    if x >= 0:
        svar = x ** 0.5
        if svar % 1 == 0: #tallet har desimaldel .0
            forklaring = "Kvadrattall (ingen desimaldel)"
        else:
            forklaring = "Ikke kvadrattall (jfr desimaler)"
    else:
        svar = None
        forklaring = "Rot av negativt tall går ikke"
    return svar, forklaring

print(rot(4))
print(rot(5))
print(rot(-4))
```

Med nøstede if-setninger er det viktig å holde tunga rett i munnen på å få innrykkene riktig:
- return må stå slik den står nå
    - med større innrykk ville den kun blitt utført for siste else
    - da ville funksjonen ikke ha returnert noe svar for x>=0

# if-elif-else i Python
- if-else passer for avgjørelser med to alternative utfall
- ofte: N ulike utfall
    - tungvint med nøsting 

```python
def forelegg(x):
    if x <= 0:
        return 0
    else:
        if x <= 5:
            return 1100
        else:
            if x <= 10:
                return 3000
            else:
                if x <= 15:
                    return 5400
                else:
                    if x <= 20:
                        return 7800
                    else:
                        if x <= 25:
                            return 12100
                        else:
                            return "See U in court!"
```

Som vi ser funker funksjonen som den skal
- viktig her at hver nye if må puttes på else-delen til den forrige
    - er kun hvis forrige feilet at vi trenger flere beslutninger
    - (hvis forrige if slo til, har vi allerede funnet riktig beløp)

Ulempe med dette oppsettet:
- blir veldig mange kodelinjer
- og veldig mange innrykk

En enklere løsning i slike tilfeller er if-elif-else
- hvor elif er en slags forkortelse for else: med påfølgende if...


```python
def forelegg(x):
    if x <= 0:
        return 0
    elif x <= 5:
        return 1100
    elif x <= 10:
        return 3000
    elif x <= 15:
        return 5400
    elif x <= 20:
        return 7800
    elif x <= 25:
        return 12100
    else: 
        return "See U in court!"
```

En __if-elif-else__-setning
- starter med if på samme måte som før
- fortsetter med en eller flere elif
- kan helt til slutt ha en else
    - utføres hvis ingen if/elif slo til
    
Nybegynnertabbe: sette testene i en if-elif-setning i gal rekkefølge
- slik at mange av alternativene ikke vil bli vurdert
- eksempel under, mislykket forsøk på samme funksjon

```python
def forelegg(x):
    if x > 0:
        return 1100
    elif x > 5:
        return 3000
    elif x > 10:
        return 5400
    elif x > 15:
        return 7800
    elif x > 20:
        return 12100
    elif x > 25:
        return "See U in court!"
```
