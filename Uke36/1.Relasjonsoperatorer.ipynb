# Relasjonsoperatorer i Python
I matematikk og naturfag
- trenger ofte å uttrykke likheter og ulikheter

Vanlig behov  også i programmering - bruker relasjonsoperatorer:

- __==__          # er lik, sjekker om to verdier er like
- __!=__          # er ulik, betyr det motsatte av ==
- __\>__          # større enn
- __\<__          # mindre enn
- __\>=__         # større eller lik
- __\<=__         # mindre eller lik

Vær spesielt obs på __==__ som sjekker at to verdier er like
- versus __=__ som er tilordningsoperatoren (gir verdi til en variabel)

For de operatorene som består av to tegn (>=, <=, !=, ==) gjelder det at
- de må skrives i den rekkefølgen
    - omvendt rekkeføgle =!, =\<, =\> gir syntaksfeil
- de må skrives kloss inntil hverandre
    - kan ikke ha mellomrom midt i operatoren
    
Som for aritmetiske operatorer er det
- valgfritt og ofte anbefalt å ha mellomrom __rundt__ operatorene

```python
# Gitt noen verdier
a = b = 2

# Dette fungerer fint
a == 2, b >= a + 1 

# Dette funker IKKE (mellomrom midt i)
a = = 2

# Dette funker IKKE (feil rekkefølge på symbol)
a =< 2
```

## Hva blir resultatet av disse uttrykkene?
Et uttrykk med en relasjonsoperator vil beregnes til en boolsk verdi
- dvs., blir enten __True__ eller __False__

Noen eksempler:

```python
# Gir verdi til noen variable
# NB: Her brukes tilordningsoperatoren =, IKKE relasjonsoperatoren ==
a = 3
aa = 3.0
r = 3.1
b = 2

# Prøver ut div. relasjonsoperatorer, viser resultatet på skjermen
print(a > b, a >= b, a == b, a != b, a == aa, a < r, aa != r)
```

Relasjonsoperatorer har __lavere presedens__ enn aritmetiske operatorer
- og tilordningsoperatoren har __enda lavere presedens__.

Innad i aritmetiske uttrykk på hver side av en relasjonsoperator
- samme presedensregler som vist tidligere.

Dvs., gitt utførelse av variabeldefinisjonene over...
- hvordan utføres kodelinje 1 nedenfor?

```python
test1 = a + b < 2 * a
# Først: regne ut a + b (som blir 5)
# Så: regne ut 2 * a (som blir 6)
# Så: gjøre sammenligningen 5 < 6 (som blir True)
# Så: gjøre tilordningen test1 = True
# Variabelen test1 her blir dermed av type bool
print(test1)
```
