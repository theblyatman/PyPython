# Logiske betingelser med and, or, not
Ofte er beslutningene for betingelser mer sammensatte.
I Python kan vi sette sammen flere betingelser med
- operatorene __and__, __or__, __not__
- samt om nødvendig parenteser

```python
a = 1
b = 2
c = 3
print(a < b and b == c)       # blir False fordi b==c er False (og begge må være True med and)
print(a < b or b == c)        # blir True, med or er det nok at en er True 
print(a < b and not b == c)   # blir True, - b==c er False men har not foran seg så blir True
```

## Presedens mellom and, or, not

Presedensen for disse operatorene
- __not__ : høyest
- __and__ : midt
- __or__ : lavest

```python
a = 1
b = 2
c = 3
print(a < b or b < c and c > a + b) # blir True fordi b < c og c > a + b er False, men a < b er True, og med or er det nok at en er True
print((a < b or b < c) and c > a + b) # blir False fordi c > a + b er False, og med and må begge være True
```

__not__ har høyere presedens enn både __and__ og __or__

Hvis vi vil at disse andre skal gjøres før __not__ må vi bruke parenteser

Eksempel:

```python
a = 1
b = 2
print(not a > 5 or b < 5)     # blir True, fordi not a > 5 blir True og b < 5 også er True        
print(not(a > 5 or b < 5))    # blir False, fordi a > 5 or b < 5 er True, og not foran gjør det til False
print(not a <= 0 and b >= 5)  # blir False, fordi not a <= 0 er True, men b >= 5 er False
print(not(a <= 0 and b >= 5)) # blir True, fordi a <= 0 and b >= 5 er False, og not foran gjør det til True
```

```python
a = 1
b = 2
c = 3
print(a < 1 and not (b == 2 and (c < 3 or c * 2 > 4))) 

# I uttrykket over må vi først evaluere den innerste parentesen, (c < 3 or c * 2 > 4) blir True
# Deretter den ytre parentesen (b == 2 and True) = True
# Deretter not True = False
# Til slutt evaluere a < 1 and False = False
```
