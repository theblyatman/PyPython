# Strengformatering i Python: f-strenger

Enklere måte: __f-strenger__
- virker både med input, print og tilordning av strengvariable
- lettere å holde kontroll på hvor det blir mellomrom og ikke
- lettere å formatere teksten

Virkemåte:
- f like foran fnutten som starter en streng...
- sier vi at hvis det kommer __{ }__ i strengen,
- skal den regne ut det som står inni der
- og sette det i strengen i stedet

- den strengen som lages, blir en helt vanlig verdi av type string
    - akkurat som andre strenger

Effekten av bokstaven f ses lett hvis vi fjerner den:
- da blir hele greia bare en streng uten variabelverdier fylt inn

Eksempel:

```python
navn = "Ada Wong"
alder = 37

print(f'Gratulerer med {alder-10} år, {navn}!')
print('Gratulerer med {alder} år, {navn}!')
```

## formatering med f-strenger - tekst
f-strenger kan enklelt formatere tekst
- særlig relevant med flere linjer info
- som ønskes pent i kolonner

Eksemplet under gjør ingen formatering, blir rotete

```python
land = ["Norge", "USA", "New Zealand"]
landkoder = ["+47", "+1", "+64"]
print(land[0], landkoder[0]) # Norge, 47
print(land[1], landkoder[1]) # USA, 1
print(land[2], landkoder[2]) # New Zealand, 64
```

Med f-streng kan vi
- sette faste bredder på ulike felt
- får penere oppsett

```python
land = ["Norge", "USA", "New Zealand"]
landkoder = ["+47", "+1", "+64"]
print(f'{land[0]:20}{landkoder[0]:>8}') # Norge, 47
print(f'{land[1]:20}{landkoder[1]:>8}') # USA, 1
print(f'{land[2]:20}{landkoder[2]:>8}') # New Zealand, 64
```

Her blir navn på land gitt en bredde på 20 tegn
- pga __:20__ bakerst i den første { }-parentesen
- blir venstrejustert (default for tekst)
    - kunne også ha lagt til < foran tallet
- landkoden blir 8 tegn bred og høyrejustert (>8)

Ruta nedenfor illustrerer venstre- (<), midt- (^) og høyrejustering (>)
- < kan droppes for venstrejustering av tekst (default)

```python
univ = "NTNU"
f'{univ:<10}', f'{univ:^10}', f'{univ:>10}'
# 'NTNU      ', '   NTNU   ', '      NTNU'
```

## formatering med f-strenger - heltall
Bredde og justering: samme måte som tekst
- for tall er høyrejustering default
    - mens midt- eller venstre må angis eksplisitt
    

```python
folketall = 5488944
f'{folketall:<11}', f'{folketall:^11}', f'{folketall:11}'
# ('5488944    ', '  5488944  ', '    5488944')
```

Når lange heltall skal skrives i programkode
- ikke lov å dele opp, f.eks. 5,488,944

Når de skal vises for brukeren (f.eks. printes på skjerm) står vi friere
- f-strenger fikser dette
- legge til symbol som skal brukes for deling
- (men dessverre kan ikke . brukes uten videre)
    - tror da at det er desimalpunktum for tallet 11

```python
f'{folketall:11,}', f'{folketall:11_}'
# '   5,488,944', '   5_488_944'
f'{folketall:11.}'
# ' 5488944.00'
```

## formatering med f-strenger - flyttall
For flyttall kan
- bredde og justering gjøres på samme måte som for streng og heltall
- i tillegg kan vi angi antall desimaler som skal vises
- hvorvidt vi vil ha vitenskapelig notasjon (e) eller fastpunktum (f)

Eksempel:

```python
x = 1115.3423950147
f'{x:12.3f}', f'{x:15.2e}', f'{x:<7.0f}', f'{x:<10.0e}'
# ('     1115.342', '     1.12e+03', '1115   ', '1.115000e+03')
```

Eksemplet over viser fire formateringer av x
- første blir 12 tegn bred totalt, 3 desimaler, fastpunktum
- andre blir 15 tegn bred, 2 desimaler, vitenskapelig notasjon
- tredje blir 7 tegn bred, 0 desimaler, fastpunktumnotasjon
    - punktum vises ikke fordi vi valgte 0 desialer
- fjerde blir 10 tegn bred, 0 desimaler, vitenskapelig notasjon

Når antall desimaler oppgis lavere enn hva tallet har
- gir dette en avrunding av tallet som vises
- opp eller ned, avhengig av vanlige IEEE avrundingsregler
- __NB:__ innholdet i variabelen endres __ikke__
    - kun visningen av tallet som er avrundet
