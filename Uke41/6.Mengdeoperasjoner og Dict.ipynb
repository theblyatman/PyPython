# Bruke dictionary, samt mengdeoperasjoner
Skal se på hvordan vi
- slår opp data i en dictionary
- gjør mengdeoperasjoner (noen enkle, noen litt mer innfløkte)

Eksempel: har en dictionary for en globetrotterklubb
- for hvert medlem: hvilke sjeldne land besøkt?

```python
globeclub = {
    "Ola" : {"Laos", "Tonga", "Samoa", "Fiji", "Palau"},
    "Nina": {"Laos", "Palau", "Nepal", "Bhutan"},
    "Jo"  : {"Oman", "Nepal", "Pakistan", "Iran", "Laos"},
    "Ada" : {"Oman", "Jemen", "Lesotho", "Laos"},
    "Ida" : {"Laos", "Palau", "Nepal"},
    "Roy" : {"Lesotho", "Jemen", "Tonga", "Fiji", "Laos"}
}
```

## La oss gjøre noen oppslag med mengdeoperasjoner

```python
# hvilke land har både "Nina" og "Ida" besøkt?
globeclub["Nina"].intersection(globeclub["Ida"])
```

```python
# hvilke land har "Nina" besøkt men ikke "Ida"?
globeclub["Nina"].difference(globeclub["Ida"])
```

```python
# hvilke land er besøkt av enten "Jo" eller "Roy" eller begge?
globeclub["Jo"].union(globeclub["Roy"])
```

```python
# er landene til ett medlem et subsett av landene til et annet?
globeclub["Ida"].issubset(globeclub["Nina"])   
```

```python
# og omvendt?
globeclub["Nina"].issubset(globeclub["Ida"])
```

```python
# skrive funksjonen for land som er dekket av minst en person i klubben
def klubbens_samlede_land(globeclub):
    land = set()                        # starter med tom mengde
    for mengde in globeclub.values():
        land = land.union(mengde)
    return land 

print(klubbens_samlede_land(globeclub))
```

```python
# skrive funksjonen for land som alle i klubben har besøkt
def land_alle_har_vært_i(globeclub):
    land = list(globeclub.values())[0]          # her trenger vi en startverdi annen enn set()
    for mengde in globeclub.values():
        land = land.intersection(mengde)
    return land

print(land_alle_har_vært_i(globeclub))
```

```python
# funksjon som sjekker om et medlem har land som bare den og ingen andre har besøkt:
# Måte 1A: bygge opp mengde som alle andre har vært i, 
# så trekke denne fra det som medlem har vært i...
def eksklusive_land_for_medlem(globeclub, medlem):
    land = set()
    for person in globeclub:
        if person != medlem:
            land = land.union(globeclub[person])
    return globeclub[medlem].difference(land)

print(eksklusive_land_for_medlem(globeclub, "Ola"))
```

```python
# funksjon som sjekker om et medlem har land som bare den og ingen andre har besøkt:
# Måte 1B: bygge opp mengde som alle andre har vært i, for hver person legge til ett og ett land med muterende add()
# så trekke denne fra det som medlem har vært i...
def eksklusive_land_for_medlem(globeclub, medlem):
    land = set()
    for person in globeclub:
        if person != medlem:
            for land_2 in globeclub[person]:
                land.add(land_2)
    return globeclub[medlem].difference(land)

print(eksklusive_land_for_medlem(globeclub, "Ola"))
```

```python
# funksjon som sjekker om et medlem har land som bare den og ingen andre har besøkt:
# Måte 2: starte med medlemmets land
# så trekke fra landene til ett og ett annet medlem
def eksklusive_land_for_medlem(globeclub, medlem):
    land = globeclub[medlem]
    for person in globeclub:
        if person != medlem:
            land = land.difference(globeclub[person])
    return land

print(eksklusive_land_for_medlem(globeclub, "Ola"))
```

```python
# funksjon som sjekker om et medlem har land som bare den og ingen andre har besøkt:
# Måte 2: starte med medlemmets land
# så trekke fra landene til ett og ett annet medlem, ett og ett land,  med muterende discard()
def eksklusive_land_for_medlem(globeclub, medlem):
    land = globeclub[medlem]
    for person in globeclub:
        if person != medlem:
            for land_2 in globeclub[person]:
                land.discard(land_2)                # Bør derimot IKKE bruke remove() her, den feiler hvis ikke fins
    return land

print(eksklusive_land_for_medlem(globeclub, "Ola"))
```
