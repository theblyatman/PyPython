# Vanlige listemetoder som er verdt å kunne
Python har mange spesialmetoder for lister
- dekker ikke alle her, men det vanligste

NB: På testen blir det tilgjengelig et vedlegg i Inspera
- med en kort forklaring av hver listemetode
- trenger derfor ikke pugge dette...

Viktigst å forstå: Forskjell på om metoden 
- muterer eller ikke
- returnerer noe eller ikke

## list.append()
Legger til et element bakerst
- muterer listeobjektet, returnerer ingenting

```python
my_list = [1,2,3]
my_list.append(7)      # Legger til tallet 7 som siste element
print(my_list)
```

```python
my_list = [1,2,3]
my_list.append([7])     # Legger til lista [7] som siste element
print(my_list)
```

```python
my_list = [1,2,3]
x = my_list.append(7)    # vi tror kanskje at x blir [1,2,3,7]
print(x, my_list)        # viser at append IKKE returnerer noe, x er None, bare my_list er [1,2,3,7]
```

```python
my_list = [1,2,3]
x = [0,0,0] + my_list.append(4)  # kunne tro at dette blir [0,0,0,1,2,3,4], men...
print(my_list)
```

## list.extend()
Legge til ei liste av element bakerst
- muterer listeobjektet, returnerer ingenting

```python
my_list = [1,2,3]
my_list.extend([5, 2])
print(my_list)
```

```python
my_list = [1,2,3]
my_list.extend(5)
print(my_list)
```

```python
my_list = [5,2,4]
my_list = my_list.extend( [1,2] )
print(my_list)
print(my_list + [7])   # nei, dette blir ikke [5,2,4,1,2,7]
```

```python
# argumentet til extend trenger ikke være liste, kan også være mengde...
my_list = [1,2,3]
my_list.extend({7,8})
my_list
```

```python
# ...eller det kan være array (eller annen samling av verdier)
import numpy as np
arr = np.array([7,8])
my_list = [1,2,3]
my_list.extend(arr)
my_list
```

## list.copy()
Returnerer en kopi av lista
- endrer ikke originalen, og kopien blir et separat listeobjekt i minnet

```python
my_list = [1,2,3]
new_list = my_list.copy()
my_list[0] = 9
print(my_list, new_list)
```

## list.clear()
Muterer lista ved å endre til tom liste

```python
my_list.clear()
my_list
```

## list.index(verdi)
Returnerer indeksposisjonen (første fra venstre) for __verdi__ i liste
- muterer IKKE
- gir ValueError hvis verdi ikke fins

```python
my_list = [1,2,3,1]
my_list.index(1)
```

```python
my_list = [1,2,3,1]
my_list.index(0)
```

## list.insert(indeks, verdi)
Muterer lista ved å skvise inn verdi i posisjonen indeks
- alle bakenforliggende element i lista flyttes én indeksposisjon for å gi plass
- returnerer ingenting

```python
my_list = [1,2,3]
my_list.insert(-100,5)
my_list
```

```python
my_list = [5,2,4,0,7]
my_list.insert(1,3)
print(my_list)
```

```python
my_list = [5,2,4]
my_list = my_list.insert(1,3)
print(my_list)
```

## list.pop(indeks)
Muterer lista ved å fjerne verdien som står i posisjon __indeks__
- bakenforliggende element flyttes ett hakk fram
- returnerer verdien som ble fjernet

```python
my_list = [5, 2, 4, 7, 0]
x = my_list.pop(0)
print(x, my_list)
```

```python
my_list = [5,2,4,7]
x = my_list.pop(4)
print(x, my_list)
```

## list.remove(verdi)
Muterer lista ved å fjerne første forekomst av __verdi__
- returnerer ingenting
- gir ValueError hvis verdi ikke fins

```python
my_list = [5,7,4,3,7]
my_list.remove(7)
my_list
```

```python
my_list = [5,2,4]
my_list.remove(0)
print(my_list)
```

## list.reverse()
Muterer lista ved å snu element i motsatt rekkefølge
- returnerer ingenting

```python
my_list = [5, 7, 1, 4, 2]
my_list.reverse()
print(my_list)
```

## list.sort()
Muterer lista ved å putte elementene i sortert rekkefølge
- default: sortert i stigende rekkefølge
- returnerer ingenting
- virker kun hvis alle element er tall, eller alle er tekst

```python
my_list = [5, 7, 1, 4, 2]
my_list.sort()
print(my_list)
```

```python
my_list = [5, 7, 1, 4, 2]
my_list = my_list.sort()  # Dette blir feil!
print(my_list)
```

```python
my_list = [5, 7, "A", 4, 2]
my_list.sort()
print(my_list)
```

### Nøkkelordargument til sort()
- Synkende rekkefølge kan oppnås med argumentet __reverse=True__
- Sortering på annet enn verdi kan oppnås med __key=...__

```python
my_list = [5,7,5,4,3,7]
my_list.sort(reverse=True)
my_list
```

```python
name_list = ["Ada", "Jo", "Ruby", "Pascal"]
name_list.sort(key=len)
name_list
```

## ListeFUNKSJONEN sorted()
I motsetning til metoden __list.sort()__
- funksjonen muterer __ikke__ original liste
- returnerer i stedet et nytt listeobjekt med elementene i sortert rekkefølge
- samme mulige nøkkelordargument som list.sort()

Siden sorted() returnerer, blir det riktig å putte resultatet inn i en variabel:

```python
my_list = [5, 7, 1, 4, 2]
sorted_list = sorted(my_list)
print(my_list, sorted_list)
```

## ListeFUNKSJONEN reversed()
I motsetning til metoden __list.reverse()__
- funksjonen muterer __ikke__ original liste
- returnerer i stedet et generator-objekt med elementene i motsatt rekkefølge

Da blir det riktig å putte resultatet inn i en variabel
- med generatorobjekt returnert må vi bruke list() i tillegg for å få ei liste

```python
import numpy as np

my_list = [5, 7, 1, 4, 2]
rev_list = reversed(my_list)
print(rev_list)
print(list(rev_list), my_list)
```

## Oppsummering
Python har mange nyttige listemetoder og -funksjoner
- metoder kalles ved __listevariabel.metodenavn()__
- funksjoner kalles ved __funksjonsnavn(listevariabel)__

Viktig å skjønne forskjell på
- metoder som muterer lista, men ikke returnerer den
- vs. metoder / funksjoner som returnerer ei liste

Hvis du bruker en metode som ikke returnerer noe, som om den returnerte...
- blir det feil - verdien er __None__, som ikke er brukbar i videre beregninger
