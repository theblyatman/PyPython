# Utvida tilordning
I mange situasjoner har vi behov for å endre verdien til en variabel, f.eks.
- øke eller redusere den med 1
- dele eller gange den med 2

Særlig i løkker forekommer slike operasjoner ofte, f.eks.
- telle noe: legge til 1 per runde
- summere noe: addere inn nytt tall per runde
- regne ut et produkt: gange inn nytt tall per runde
- ...

For å kunne skrive slike operasjoner mer kompakt
- har vi utvida tilordningsoperatorer.

Merk at med utvidet tilordning __MÅ__ variabel på venstre side være definert fra før:
- f.eks. med en vanlig tilordning tidligere i koden
- fordi man trenger gammel verdi av variabel for å regne ut ny

```python
# FLERE EKSEMPLER, -, *, +
a = b = c = d = 3
a -= 2       # samme som a = a - 2
b *= 4       # samme som b = b * 4
c += d       # samme som c = c + d
print(a, b, c)

# FLERE EKSEMPLER: /, //, %
x = y = z = 15
x /= 4          # samme som x = x / 4
y //= 4         # samme som y = y // 4
z %= 4          # samme som z = z % 4
print(x, y, z)
```

## Mer detaljer om virkemåte
For godt til å være sant:
- "Utvida tilordning er __synonymt__ med vanlig tilordning" (celle 5, 6 over)

For sant til å være godt:
- dette stemmer __kun__ for immuterbare datatyper (f.eks. int, float, bool, string, tuple)
- stemmer __ikke__ for muterbare datatyper (f.eks. list)

```python
liste = [1, 2, 3]
print(id(liste))
liste += [4]   # += gjør endringen i __samme listeobjekt__
print(liste)
print(id(liste))  # vi ser dette ved at id() ikke har endret seg
                  # liste-objektet er fortsatt på samme minneadresse
```

```python
liste = [1, 2, 3]
print(id(liste))
liste = liste + [4]   # = ... + lager et __nytt listeobjekt__
print(liste)
print(id(liste))  # vi ser dette ved at id() har endret seg
                  # i linje 5 har liste-objektet annen minneadresse enn i linje 2
```

Det samme gjelder __liste \*= n__  versus __liste = liste \* n__
- første repeterer innholdet __i__ samme listeobjekt
- andre lager nytt listeobjekt

Implikasjon: for muterbare datatyper som liste må du holde tunga rett i munnen...

```python
mine_favorittlag = ["Florø", "Burnley", "Palermo"]
vennens_favorittlag = mine_favorittlag
vennens_favorittlag = vennens_favorittlag + ["Al-Nassr"]
print(vennens_favorittlag) # ['Florø', 'Burnley', 'Palermo', 'Al-Nassr']
print(mine_favorittlag) # ['Florø', 'Burnley', 'Palermo']
```
