# Lokale og globale variable
Variable kan ha ulik synlighet i programmet

__Global variabel__: en variabel som er definert i skriptet (hovedprogrammet) er synlig i hele programmet
- eksemplet under: det går fint å printe k inni funksjonen f
- og selvsagt også i skriptet

__Lokal variabel__: en variabel som er definert inni en funksjon, er synlig kun i denne funksjonen (og evt. hjelpefunksjoner __inni__ denne funksjonen igjen)
- vi kan printe x og y inni f
- men IKKE i skriptet

```python
# Funksjonen f
def f(x):
    y = 2
    print(f'x inni f: {x}')
    print(f'y inni f: {y}')
    print(f'k inni f: {k}')

# Skriptet (hovedprogram)
k = 3
print(f'k i skriptet: {k}') 
f(5)
print(f'y i skriptet: {y}')
```

## Samme navn kan brukes i flere skop
- og da potensielt vise til ulike objekter

Dette er nyttig i mange situasjoner
- akkurat som mange byer kan ha Strandgata, Kongens gate osv.
- vi kan bruke fornuftige navn inni en funksjon
    - uten å måtte tenke på om navnet er brukt et eller annet sted i programmet

```python
def f(x):
    y = 7
    print(f'y verdi {y} og minnelokasjon {id(y)} inni f')

# skriptet
y = 4
f(9)
print(f'y verdi {y} og minnelokasjon {id(y)} i skriptet')
```

Hva om vi ønsket at
- y inni f skal være samme variabel som i skriptet?
- må bruke ordet __global__ foran
    - vanligvis en dårlig måte å programmere på

```python
def f(x):
    global y
    y = 7
    print(f'y verdi {y} og minnelokasjon {id(y)} inni f')

# skriptet
y = 4
f(9)
print(f'y verdi {y} og minnelokasjon {id(y)} i skriptet')
```
