Til slutt: Prøve å gjøre litt mer ut av det
- plotte med ulike farger og linjetyper
    - tredje parameter (streng) etter x- og y-verdier
    - ulike farger, b (blå), g (grønn), k (svart), r (rød), m (magenta), ...
    - ulike linjetyper, : prikker, -- streker, o rundinger, + plusser, ...

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 - 3*x - 1

def g(x):
    return 3 * np.sin(x)

x = np.linspace(-2, 4, 51)

fig, ax = plt.subplots()
ax.plot(x, f(x), "g--")
ax.plot(x, g(x), "m:")
ax.legend(["f", "g"])
plt.title("f og g som funksjoner av x")
plt.xlabel("x-verdi")
plt.ylabel("y-verdi")
plt.show()
```

Og helt til slutt: 
- få til aksekors i origo
- fant kode for dette på StackOverflow

https://stackoverflow.com/questions/31556446/how-to-draw-axis-in-the-middle-of-the-figure

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 - 3*x - 1

def g(x):
    return 3 * np.sin(x)

x = np.linspace(-2, 4, 51)

fig, ax = plt.subplots()
ax.plot(x, f(x), "g--")
ax.plot(x, g(x), "m:")
ax.legend(["f", "g"])

# DENNE KODEN FOR Å FÅ TIL AKSEKORSET
# Flytte venstre strek og bunnstrek til 0, 0
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Skjule høyre strek og toppstrek
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# Bare vise haker ved venstre og bunn
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.title("f og g som funksjoner av x")
plt.xlabel("x-verdi")
plt.ylabel("y-verdi")
plt.show()
```

## Plotte kakediagram med
- Det er mulig å plotte uten å lage en funksjon for det
- Fordel med funksjon:
  - kan bruke samme kode om igjen flere sted i programmet
  - slipper å huske detaljer om hvordan plottingen kodes

```python
import matplotlib.pyplot as plt
import numpy as np

def plott_kake(tall, tekster, farger, tittel):
    '''
    Plotter et kakediagram.
    Parameters
    ----------
    tall : array eller liste med tall (int eller float)
        tallene som bestemmer relativ størrelse på kakestykkene, f.eks 50, 34, 16
    tekster : array eller liste med string
        navnene på kakestykkene.
    farger : array eller liste med string
        farger som skal gis til kakestykkene (må gis på engelsk, små bokst.)
    tittel : string
        en tittel for hele diagrammet.

    Returns
    -------
    None.

    '''
    fig, ax = plt.subplots()
    ax.pie(tall, labels=tekster, colors=farger, autopct='%1.0f%%')
    plt.title(tittel)
    plt.show()
```

Vi kan nå bruke funksjonen ved å kalle den:

```python
pelsfarge = ["Flekket", "Svart"]
andel_farger = np.array([89, 11])
farger = ["orange", "grey"]
plott_kake(andel_farger, pelsfarge, farger, "Andel flekkede vs. svarte jaguarer")
```

Annen måte:

```python
import matplotlib.pyplot as plt
import numpy as np

def plott_kake(tall, tekster, farger, tittel, prosent):
    '''
    Plotter et kakediagram.
    Parameters
    ----------
    tall : array eller liste med tall (int eller float)
        tallene som bestemmer størrelse på kakestykkene, f.eks 50, 34, 16
        for prosentandeler av ja / nei / vet ikke.
    tekster : array eller liste med string
        navnene på kakestykkene.
    farger : array eller liste med string
        farger som skal gis til kakestykkene (må gis på engelsk, små bokst)
    tittel : string
        en tittel for hele diagrammet.
    prosent : string
        tekst for å angi prosentnotasjon i kakestykkene

    Returns
    -------
    None.

    '''
    fig, ax = plt.subplots()
    ax.pie(tall, labels=tekster, colors=farger, autopct=prosent)
    plt.title(tittel)
    plt.show()

plott_kake([138,125,70], ["Ja", "Nei", "Avstår"], None, "Avstemningsresultat", '%1.1f%%')
```

## Plotte søylediagram med Python / Matplotlib
- Det er mulig å plotte uten å lage en funksjon for det
- Fordel med funksjon:
  - kan bruke samme kode om igjen flere sted i programmet
  - slipper å huske detaljer om hvordan plottingen kodes

```python
import matplotlib.pyplot as plt
import numpy as np

def plott_søyle(tekster, tall, tittel):
    '''
    Plotter et søylediagram.
    Parametre (tekster og tall må ha like mange element)
    ----------------------------------------------------------------
    tekster : array eller liste med string
        Gir tekstetikettene som tilhører hver enkelt søyle.
    tall : array eller liste med int eller float
        Gir tallene som er høyde på søylene.
    tittel : string
        Gir en tittel for hele diagrammet.

    Returns
    -------
    None.

    '''
    fig, ax = plt.subplots()
    ax.bar(tekster, tall)
    plt.title(tittel)
    plt.show()

land = ["Brasil", "Peru", "Colombia", "USA", "Venezuela"]
jaguarer_per_land = np.array([86800, 22000, 15000, 15000, 8600])
plott_søyle(land, jaguarer_per_land, "Landene med flest jaguarer")
```

Kan lage tilsvarende funksjon for horisontalt søylediagram
- eller la oss kalle det et bjelkediagram...

```python
def plott_bjelke(tekster, tall, tittel):
    fig, ax = plt.subplots()
    ax.barh(tekster, tall)
    plt.title(tittel)
    plt.show()

plott_bjelke(land, jaguarer_per_land, "Landene med flest jaguarer")
```

## H-11c: Plotte bilder
- importere matplotlib og numpy
- importere PIL (Python Imaging Library)
- viser kode tilsvarende H-testen:

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def plott_bilde(filnavn, tittel):
    fig, ax = plt.subplots()
    img = np.asarray(Image.open(filnavn))
    ax.tick_params(left = False, right = False, labelleft = False, 
                   labelbottom = False, bottom = False)
    plt.title(tittel)
    plt.imshow(img)


def main():
    print("Gauk (Cuculus canorus) er ein reindyrka reisparasitt")
    print("Byggjer ikkje eige reir, legg egga sine i andre reir")
    plott_bilde('./gauk.png', 'Gauk (Cuculus canorus)') # "./<...>" leter etter et bilde i samme mappe som pythonfilen som kjøres.

    # ...
    print("Gulnebbgauk (Coccyzus americanus) byggjer reir og ruger sjølv,")
    print("men kan av og til også leggje egg i andre reir.")
    plott_bilde('./gulnebbgauk.png', 'Gulnebbgauk (Coccyzus americanus)')

main()
```
