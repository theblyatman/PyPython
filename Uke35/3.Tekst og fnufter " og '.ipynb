# Tekststrenger i Python
I et Python-program er det viktig å skille mellom
- hva som er tekstverdier
- hva som er objektnavn og spesielle ord i programmeringsspråket

__I Python gjøres dette ved å omslutte tekst med fnutter__
- uten fnutter blir det tolket som forsøk på et navn

## Ulike typer fnutter rundt tekststrenger i Python
- 'apostrof' 
- "hermetegn"  
- '''trippel apostrof'''
- """trippel hermetegn"""


```python

'Python' # Disse to blir akkurat samme tekst
"Python"
# Poenget med å tilby to ulike symbol, er at tekst kan inneholde symbolet
"OK, I won't." # OK, I won't.

"Please never invite "Al" again!" # Invalid syntax
'OK, I won't.                     #  Invalid syntax

# Triple fnutter blir nødvendig hvis vi har begge typer fnutter i teksten
'''Please don't invite "Al" again, dear!''' # Please don't invite "Al" again, dear!

# Med to typer triple fnutter blir vi i stand til å nevne triple fnutter i teksten:
"""Flerlinjes strenger i Python kan omgis av '''triple
fnutter'''""" 
# Flerlinjes strenger i Python kan omgis av '''triplefnutter'''
```

```python
# Triple fnutter kan også brukes når vi har strenger over flere linjer:
'''våren kjem til den vesle hytta
ingenting er her
alt er her
Sodo (1641-1716)'''
```

## Oppsummering
- tekststrenger i Python omsluttes av fnutter
- for at de ikke skal bli mistolket som navn
- har to typer fnutter, " og '
    - muliggjør tekst med fnutter inni
- triple fnutter kan også brukes, nyttig for
    - tekst over flere linjer
    - tekst som inneholder begge typer fnutter
